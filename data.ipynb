Vérification si NA valeurs

```python
data['2003'].isnull().sum()
```

Aucun temps sur les points de passage

La feature Unnamed: 0 est inutile

```python
data['2003'] = data['2003'].drop(['Unnamed: 0'], axis=1)
```

```python
def chronoSecond(chrono_string) : 

    hours = int(chrono_string[0:2])
    minutes = int(chrono_string[3:5])
    seconds = int(chrono_string[6:8])
    
    return 3600*hours + 60 * minutes + seconds
```

On doit vérifier que les données de chrono sont toutes au format 'HH:MM:SS'

```python
print('Données au mauvais format :')
print(data['2003'].shape[0] - data['2003']['time'].str.contains('^[0-9]{2}\:[0-9]{2}\:[0-9]{2}$').sum() - 
      data['2003']['time'].isna().sum() )
```

Notons également que la feature timediff n'est pas correcte. On convertit les données de chrono en seconde

```python
data['2003']['time'] = data['2003']['time'].apply(lambda x: chronoSecond(x))
```

```python
time_winner = data['2003'].time[0]
data['2003']['timediff'] = data['2003']['time'] - time_winner 
```

Points de passage sur les éditions 2004 à 2017 en plus, mais ces points de passage sont différents selon les éditions

On remarque que les catégories sont définies différemment entre 2003 et les autres éditions. Par exemple SEF en 2003 devient SE F sur les autres éditions. On va reprendre le modèle de 2003 pour toutes les éditions

```python
for y in list(data.keys())[1:] :
    data[y]['category'] = data[y]['category'].apply(lambda x: re.sub('\s', '',x))
```

On vérifie s'il y a des données manquantes. Une donnée manquante peut être formalisée sous la forme d'un NaN ou d'un string vide

```python
data['2004'].isna().sum()
```

```python
(data['2004'] == ' ').sum()
```

Des sont manquants pour le chrono final et certains intermédiaires. Soit c'est un disfonctionnement, soit c'est un abandon. Dans le cas où la donnée est un string vide, on la formalise par un NaN

De même, suppression de la colonne Unnamed: 0, de colonne timediff et conversion des string vides en NaN

```python
for y in list(data.keys())[1:] :
    data[y] = data[y].drop(['Unnamed: 0', 'timediff'], axis = 1)
    data[y] = data[y].replace(' ', float('nan'))
```

```python
data['2004']
```

On vérifie que les données de chrono sont au bon format

```python
for y in list(data.keys())[1:] :
    
    cols = data[y].columns[6:]
    for c in cols :
        print('Données au mauvais format année {} feature col {} :'.format(y, c))
        if data[y][c].isna().sum() == data[y][c].shape[0] :
            print('Only NaN : bug chrono')
        else :
            print(data[y].shape[0] - data[y][c].str.contains('^[0-9]{2}\:[0-9]{2}\:[0-9]{2}$').sum() - 
                  data[y][c].isna().sum() )    
```

De nombreuses données ne sont pas au bon format, il faut analyser ces erreurs

```python
for y in list(data.keys())[1:] :
    
    cols = data[y].columns[6:]
    for c in cols :
        for i in range(data[y].shape[0]) :
            if type(data[y].loc[i, c]) == str :
                if re.search('^[0-9]{2}\:[0-9]{2}\:[0-9]{2}$', data[y].loc[i, c]) == None :
                    print('année {}, feature {}, indice {}, string {}'.format(y, c, i, data[y].loc[i, c]))
```

On remarque que les erreurs sont toujours de même type, un - est ajouté devant le chrono.
On supprime ce - sur toutes les tables

```python
for y in list(data.keys())[1:] :
    cols = data[y].columns[6:]
    for c in cols :
        data[y][c] = data[y][c].apply(lambda x : re.sub('\-', '', x) if type(x) == str else x)
```

On peut vérifier qu'il n'y a plus d'erreur

```python
for y in list(data.keys())[1:] :
    
    cols = data[y].columns[6:]
    for c in cols :
        for i in range(data[y].shape[0]) :
            if type(data[y].loc[i, c]) == str :
                if re.search('^[0-9]{2}\:[0-9]{2}\:[0-9]{2}$', data[y].loc[i, c]) == None :
                    print('année {}, feature {}, indice {}, string {}'.format(y, c, i, data[y].loc[i, c]))
```

C'est okay, il n'y a plus d'erreur qui s'affiche

On peut désormais convertir les données de chrono en seconde convenablement et calculer la feature timediff

```python
for y in list(data.keys())[1:] :
    
    cols = data[y].columns[6:]
    for c in cols :
        data[y][c] = data[y][c].apply(lambda x : chronoSecond(x) if type(x) == str else x)
    
    time_winner = data[y].time[0]
    data[y]['timediff'] = data[y]['time'] - time_winner
```

### Export Data

On peut désormais exporter ces tables nettoyées dans un répertoire dédié

```python
for y in data.keys() :
    file_path = y+'-utmb.csv'
    path = os.path.join('clean_data', file_path)
    data[y].to_csv(path)
```
